# Notebook to merge and clean logfiles

When Basilisk simulations exceed the cluster walltime, the log-information will be spread over multiple files. Sometimes the logs also contain error messages coming from the cluster's operating system. Moreover, the restart of a simulation leads to duplicated data in the logs since restart files are not written after each iteration. The aim of this notebook is to

- read the log files for all simulations
- select only files which actually contain sensible logs and remove erroneous lines
- save the cleaned and merged logs for each simulation
- read the new logs and display the terminal velocity as a first sanity check

```python
from glob import glob

# create paths to folders
source_base = "../data/"
target_base = "../data/"

case_list = ["dell_l15", "dell_l16", "dell_l17", "scap_l15", "scap_l16", "wa18_l15", "wa18_l16"]
log_dict = {}
for case in case_list:
    log_dict[case] = sorted(glob(source_base + case + "/{:s}*".format(case)))
log_dict
```

```python
import pandas as pd


def read_and_clean_logs(logs):
    '''Read single log files from disk, clean, and merge.
    
    Parameters
    ----------
    logs - array-like : List with paths to log files
    
    Returns
    -------
    full_log - DataFrame : merged log file as Pandas dataframe
    
    '''
    names = ['time', 'volume', 'x', 'y', 'z', 'u_x', 'u_y', 'u_z', 'area']
    log_data = []
    for file in logs:
        raw_log = pd.read_csv(file, sep=' ', header=None, names=names, low_memory=False)
        raw_log.dropna(inplace=True)
        clean_log = raw_log[pd.to_numeric(raw_log.time, errors='coerce').isnull() == False].reset_index(drop=True)
        log_data.append(clean_log.apply(pd.to_numeric))
        
    if (len(log_data) > 1):
        for l in range(len(log_data) - 1):
            start = log_data[l + 1].time.values[0]
            log_data[l] = log_data[l][log_data[l].time < start]
    full_log = pd.concat(log_data)
    return full_log.reset_index(drop=True)
```

## Load, clean, merge and save log files

```python
for case in case_list:
    print("Processing case {}".format(case))
    log = read_and_clean_logs(log_dict[case])
    print(log.shape)
    target = target_base + "/" + case + "/log." + case + ".csv"
    log.to_csv(target, sep=',', index=False)
    print("-" * 20)
```

## Load clean logs and plot colums

```python
col_names = ['time', 'volume', 'x', 'y', 'z', 'u_x', 'u_y', 'u_z', 'area']
clean_logs = {}
for case in case_list:
    log = target_base + "/" + case + "/log." + case + ".csv"
    clean_logs[case] = pd.read_csv(log)
```

### Velocity components

```python
# execute twice
for case in clean_logs.keys():
    clean_logs[case].plot('time', ['u_x','u_y'], title=case, figsize=(16, 4), xlim=(0, 50))
```
